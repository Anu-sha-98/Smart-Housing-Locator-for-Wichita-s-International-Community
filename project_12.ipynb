```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset into a pandas DataFrame
df = pd.read_csv('C:/Users/alish/OneDrive/Documents/Project/Project_Dataset.csv')

# Extract 'Monthly_Rent' variable
monthly_rent = df['Monthly_Rent']

# Set up the figure and axes
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the histogram for 'Monthly_Rent'
ax.hist(monthly_rent, bins=20, color='skyblue', edgecolor='black', alpha=0.7, density=False)

# Set labels and title
ax.set_title('Histogram of Monthly Rent with Frequency Counts')
ax.set_xlabel('Monthly Rent')
ax.set_ylabel('Frequency')

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset into a pandas DataFrame
df = pd.read_csv('C:/Users/alish/OneDrive/Documents/Project/Project_Dataset.csv')

# Extract 'Management_Efficiency' variable
management_efficiency = df['Management_Efficiency']

# Calculate frequency counts
frequency_counts = management_efficiency.value_counts()

# Set up the figure and axes
fig, ax = plt.subplots(figsize=(8, 5))

# Plot the bar plot for 'Management_Efficiency' frequency counts
frequency_counts.plot(kind='bar', color='skyblue', edgecolor='black', alpha=0.7)

# Set labels and title
ax.set_title('Frequency of Management Efficiency Ratings')
ax.set_xlabel('Management Efficiency')
ax.set_ylabel('Frequency')

# Display the plot
plt.show()
```
